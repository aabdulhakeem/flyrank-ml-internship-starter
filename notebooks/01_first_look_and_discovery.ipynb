## 2. Discover a real truth yourself

The safest, most satisfying early wins are **things you find in the data** — un-leakable, and they *are* the core lesson. Run the three cells below. Each is ~10 lines of pandas and each overturns a common SEO belief.

Every number is **computed live from the shipped CSV** — nothing is hardcoded.

```python
import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(df.shape[0], "rows,", df.shape[1], "columns")
df.head(3)
```

### Discovery A — "High search volume means more traffic." Does it?

```python
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")
print("Near zero -> keyword search volume barely predicts the traffic a page actually gets.")
```

### Discovery B — the CTR cliff by position
Click-through rate is not flat: it collapses as you move down the results.

```python
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos.round(4).to_string())
ctr_by_pos.plot(kind="bar", title="Mean CTR by position tier (impressions >= 100)", ylabel="CTR");
```

### Discovery C — is longer content the lever?
Compare word count for **declining** vs **growing** pages.

```python
wc = df.groupby("trend_direction")["word_count"].median()
print(wc.round(0).to_string())
print("\n'down' vs 'up' pages have almost the same median word count -> length is not the lever.")
```

```python
# Discovery A but only for pages with impressions_90d > 0 — does the correlation change?
df[df['impressions_90d'] <= 0]['impressions_90d'].count()
```

* Observed: there are no rows in this slice that have impressions_90d  <= 0 so filtering for impressions_90d > 0 doesn't change the sample at all — the correlation stays the same.
* This suggests the near-zero correlation in Discovery A isn't an artifact of zero-impression rows; it appears to be a real, weak relationship in this slice.

```python
# Discovery B, which `content_type` has the worst CTR *at the same position tier*?
import seaborn as sns
import matplotlib.pyplot as plt

visible = df[df["impressions_90d"] >= 100]

tier_order = ["top_3", "page_1", "striking", "page_3_5", "deep", "no_data"]

plt.figure(figsize=(10, 6))
sns.barplot(
    data=visible,
    x="position_tier",
    y="ctr",
    hue="content_type",
    order=tier_order,
    estimator="mean",
    errorbar=None
)
plt.ylabel("Mean CTR")
plt.title("Mean CTR by position tier and content_type (impressions_90d >= 100)")
plt.show()
```

```python
counts = visible.groupby(["position_tier", "content_type"])["ctr"].count().unstack()
print(counts)
```

* Observed: after filtering to impressions_90d >= 100, `keyword article` is the only
  content_type with a reliable sample size across all position tiers (527–8,186 rows).
* `comparison article` and `feedly article` have very small samples in `top_3` and `deep`
  specifically — comparison article has just 1 row in top_3 and 0 rows in deep; feedly
  article has only 5 rows in top_3 and 10 in deep. Any CTR comparison in these two tiers
  is not statistically meaningful for these content types (a single click could swing the
  rate by tens of percentage points).
* Within tiers with reasonable sample size (page_1, striking, page_3_5), comparison article
  shows a consistently lower mean CTR than keyword article — this looks like a real, directional
  pattern worth investigating further, not noise.
* feedly article's apparent CTR advantage in top_3/deep is likely an artifact of tiny sample
  size and should not be reported as a finding without a larger slice or the full warehouse data.


```python
# does `content_age_days` relate to `trend_direction`?
age = df.groupby('trend_direction')["content_age_days"].mean().sort_values(ascending=False)
age.plot(kind="bar", title="Mean age by trend direction", ylabel="age")
```

```python
update_by_trend = df.groupby("trend_direction")["days_since_last_update"].mean().sort_values(ascending=False)
print(update_by_trend)
```

* Observed: mean content_age_days is higher for stable/up (~290-295) vs down/flat/new
  (~235-245), a ~60 day gap.
* Tested a possible explanation: hypothesized that stable pages might simply have been
  refreshed more recently than down pages (days_since_last_update as a confound). This did
  NOT hold — stable (50.1 days) and down (49.2 days) have nearly identical mean
  days_since_last_update, ruling out "recency of update" as the explanation for the age gap.
* The age difference between trend_direction groups remains unexplained by update recency;
  worth investigating other factors (content_type, keyword competitiveness) before drawing
  conclusions.
